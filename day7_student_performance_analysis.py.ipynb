```python
score=[
    df["math score"].mean(),
    df["reading score"].mean(),
    df["writing score"].mean(),
]
subject=["math","reading","writing"]
plt.bar(subject,score)
plt.title("Average student performance score")
plt.xlabel("subject")
plt.ylabel("score")
plt.show()
```

```python
df.groupby("test preparation course")[["math score","reading score","writing score"]].mean()
```

```python
df.groupby("parental level of education")[["math score","reading score","writing score"]].mean().sort_values("math score")
```

```python
import matplotlib.pyplot as plt

df.groupby("gender")["math score"].mean().plot(kind="bar")
plt.title("Average math score by gender")
plt.ylabel("Average math score")
plt.show()
```

```python
df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean().plot(kind="bar")
plt.title("Impact of Test Preparation Course on Scores")
plt.ylabel("Average Score")
plt.show()
```
